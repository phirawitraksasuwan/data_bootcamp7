## Filter dataframe more than one condition
- and &
- or |

```python
# Filter more than one condition (and &, or |)
filter_penguins = penguins[ ( penguins['island'] == 'Torgersen' ) & ( penguins['bill_length_mm'] < 35 ) ]
filter_penguins
```

## Missing Values

```python
# check missing in each column
penguins.isna()
```

```python
# sum missing value
penguins.isna().sum()
```

```python
# filter missing value in column sex
penguins['sex'].isna()
penguins[penguins['sex'].isna()]
# ผลลัพธ์ จะมีค่า missing value ที่ column sex 11 row
```

## Drop na 

```python
# drop na 
clean_penguins = penguins.dropna()
clean_penguins.head(10)
# ผลลัพธ์จะไ่ม่มีค่า nan หรือ missing value
```

## Fill missing values

```python
# จะเเทนที่ค่าเฉลี่ยของ column นั้นๆ ลงใน missing value 
# ตัวอย่างที่ ' bill_length_mm '

# fill missing value
top5_penguins = penguins.head(5)
top5_penguins
```

```python
top5_penguins['bill_length_mm'].mean()
avg_value = top5_penguins['bill_length_mm'].mean()
print(avg_value)
```

```python
# fillna() เป็น methods เอาไว้เเทนที่ค่่า nan
top5_penguins['bill_length_mm'].fillna(value=avg_value)
```

## ลองจัดการกับ missing value เอง โดย mean lmputation

```python
# ลองทำเอง
column = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']

for col in column :
    avg = penguins[col].mean()
    penguins[col] = penguins[col].fillna(avg)

# col = ตัวแปร จะเป็นอะไรก็ได้ เเต่ต้องสัมพันกับสูตรการคำนวน คือต้องเอามาคิดได้ 
# column = เป็นเงื่อนไขว่า จุดเริ่มต้นกับจุดสิ้นสุดคือเท่าไหร่
penguins
```
