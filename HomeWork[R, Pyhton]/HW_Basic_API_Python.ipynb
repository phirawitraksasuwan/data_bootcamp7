# API 2

```python
url = "https://www.gov.uk/bank-holidays.json"
result = requests.get(url)

data = result.json()
data
```

```python
data["england-and-wales"]["events"][66]
```

```python
from requests import get

nums = list(range(0,66))
result = []

for i in nums :
    url = "https://www.gov.uk/bank-holidays.json"
    response = get(url)
    data = response.json()
    event = data["england-and-wales"]["events"][i]
    row = [
        event["title"],
        event["date"],
        event["notes"],
        event["bunting"]
    ]
    result.append(row)
print(result)
```

```python
import pandas as pd
df = pd.DataFrame(result)
df.columns = ["Title","Date","Notes","Bunting"]
df
```

# API 3

```python
url = "https://www.balldontlie.io/api/v1/players"
result = requests.get(url)

result.json()
```

```python
from requests import get
nums = list(range(1,51))
result = []

for i in nums:
    url = f"https://www.balldontlie.io/api/v1/players/{i}"
    response = get(url)
    data = response.json()
    row = [
        data["first_name"],
        data["last_name"],
        data["height_feet"],
        data["height_inches"],
        data["position"],
        data["team"]["full_name"]
    ]
    result.append(row)

print(result)
```

```python
import pandas as pd
df = pd.DataFrame(result)
df
```

```python
df.columns = ["first_name", "last_name", "height_feet", "height_inches", "position", "team"]
```

```python
df
```

```python
import requests

nums = list(range(1, 26))
result = []

for i in nums:
    url = f"https://api.punkapi.com/v2/beers/{i}"
    response = requests.get(url)
    data = response.json()
    row = [
        data[0]["id"],
        data[0]["name"],
        data[0]["ph"]
    ]
    result.append(row)

print(result)
```

```python
import pandas as pd

df = pd.DataFrame(result)
df.columns = ["ID","Name","ph"]
df
```
